## Averias
Hem suposat que quan el vehcle es troba a velocitat zero, està atenent una averia

```python
import datetime
ubicaciones = pd.read_csv('../datasets/ubicaciones.csv', sep=';')
ubicaciones.data_hora = pd.to_datetime(ubicaciones.data_hora, format = '%Y-%m-%dT%H:%M:%S')
limit = datetime.timedelta(minutes=15)
averias = pd.DataFrame()

# Per franja horaria
for i in range(0,len(ubicaciones.data_hora)-2):
    if (ubicaciones.data_hora[i+1] - ubicaciones.data_hora[i]> limit):
        averias = pd.concat([averias, ubicaciones.loc[[i]]])

# Per velocitat
averias1 = ubicaciones[ubicaciones.velocitat == 0]
averias1.to_csv('averias.csv', sep=";")
```

## Baixes

```python
bajas = pd.read_csv('../datasets/Bajas.csv', sep=";")
bajas.columns = ['Mun','any','mes','dia','censal','uso_agua','vivienda','numero_contratos']
bajas = pd.DataFrame(bajas)
def str_join(df, sep, *cols):
    from functools import reduce
    return reduce(lambda x, y: x.astype(str).str.cat(y.astype(str), sep=sep),[df[col] for col in cols])

bajas['date'] = str_join(bajas, '/', 'any', 'mes', 'dia')
bajas.date = pd.to_datetime(bajas.date, format = '%Y/%m/%d')
bajas.to_csv('baixes.csv', sep=";")
```
