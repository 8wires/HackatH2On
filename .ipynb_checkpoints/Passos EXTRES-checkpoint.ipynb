## Creació del dataframe: reclamaciones

```python
import csv
import pandas as pd
requerimientos = pd.read_csv('../datasets/Hackathon_2017_DataSets_20170503/requerimientos 2010 a 2016 v20170427.csv', sep=';')
reclamaciones = requerimientos[requerimientos.Categoria == requerimientos.Categoria[1]]
reclamaciones.to_csv('reclamaciones.csv',sep=';')
```

## Averias
Hem suposat que quan el vehcle es troba a velocitat zero, està atenent una averia

```python
import datetime
ubicaciones = pd.read_csv('../datasets/ubicaciones.csv', sep=';')
ubicaciones.data_hora = pd.to_datetime(ubicaciones.data_hora, format = '%Y-%m-%dT%H:%M:%S')
limit = datetime.timedelta(minutes=15)
averias = pd.DataFrame()

# Per franja horaria
for i in range(0,len(ubicaciones.data_hora)-2):
    if (ubicaciones.data_hora[i+1] - ubicaciones.data_hora[i]> limit):
        averias = pd.concat([averias, ubicaciones.loc[[i]]])

# Per velocitat
averias1 = ubicaciones[ubicaciones.velocitat == 0]
averias1.to_csv('averias.csv', sep=";")
```

```python
import geopy
from pygeocoder import Geocoder
results = Geocoder.reverse_geocode(averias.lat.iloc[1], averias.lon.iloc[1])
```

## Baixes

```python
bajas = pd.read_csv('../datasets/Bajas.csv', sep=";")
bajas.columns = ['Mun','any','mes','dia','censal','uso_agua','vivienda','numero_contratos']
bajas = pd.DataFrame(bajas)
def str_join(df, sep, *cols):
    from functools import reduce
    return reduce(lambda x, y: x.astype(str).str.cat(y.astype(str), sep=sep),[df[col] for col in cols])

bajas['date'] = str_join(bajas, '/', 'any', 'mes', 'dia')
bajas.date = pd.to_datetime(bajas.date, format = '%Y/%m/%d')
bajas.to_csv('baixes.csv', sep=";")
```

## Contactos Solidaridad y Bajas

```python
contactos = pd.read_csv('../datasets/Hackathon_2017_DataSets_20170503/Contactos 2010 a 2016 v20170427.csv', sep =";")
solidaridad = contactos[(contactos.Motivo == "Solicitud Fondo solidaridad") | (contactos.Motivo == "Consulta Fondo solidaridad")]
solidaridad.to_csv('solidaridad.csv', sep=";")
contacto_baja = contactos[contactos.Motivo == "Baja FSP"]
contacto_baja.to_csv('contacto_baja.csv', sep = ';')
```
